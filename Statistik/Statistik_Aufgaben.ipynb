## Aufgabe 1: Stetige Gleichverteilung & Bedingte Wahrscheinlichkeit
Sei $ X $ gleichverteilt auf $[0, 1]$. Finde die Dichtefunktion (PDF) von $ Y = \sqrt{X} $. Beachte, dass $ Y $ Werte zwischen 0 und 1 annimmt. Für jedes $ y \in [0, 1] $ gilt:

$
F_Y(y) = P(Y \leq y) = P(\sqrt{X} \leq y) = P(X \leq y^2) = y^2, \quad 0 \leq y \leq 1.
$

Wir differenzieren dann und erhalten:

$
f_Y(y) = \frac{dF_Y(y)}{dy} = \frac{d(y^2)}{dy} = 2y, \quad 0 \leq y \leq 1.
$

Außerhalb des Bereichs $[0, 1]$ ist die Verteilungsfunktion $ F_Y(y) $ konstant, mit $ F_Y(y) = 0 $ für $ y \leq 0 $ und $ F_Y(y) = 1 $ für $ y \geq 1 $. Durch Differenzieren erhalten wir, dass $ f_Y(y) = 0 $ für $ y $ außerhalb $[0, 1]$.
## Lösung:

### (a)
Wir haben $ f_Y(y) = \frac{1}{l} $, für $ 0 \leq y \leq l $. Ferner, gegeben $ y $, ist die Zufallsvariable $ X $ gleichverteilt im Intervall $[0, y]$. Daher ist $ f_{X|Y}(x|y) = \frac{1}{y} $ für $ 0 \leq x \leq y $. Daraus schließen wir:

$
f_{X,Y}(x,y) = f_Y(y)f_{X|Y}(x|y) = 
\begin{cases}
\frac{1}{l} \cdot \frac{1}{y}, & 0 \leq x \leq y \leq l, \\
0, & \text{sonst}.
\end{cases}
$

### (b)
Wir haben:

$
f_X(x) = \int_x^l f_{X,Y}(x,y) dy = \int_x^l \frac{1}{l} \cdot \frac{1}{y} dy = \frac{1}{l} \ln(l/x), \quad 0 \leq x \leq l.
$

### (c)
Wir haben:

$
\mathbb{E}[X] = \int_0^l x f_X(x) dx = \int_0^l x \frac{1}{l} \ln(l/x) dx = \frac{l}{4}.
$

### (d)
Der Anteil $ Y/l $ des Stabs, der nach dem ersten Bruch übrig bleibt, und der weitere Anteil $ X/Y $ des Stabs, der nach dem zweiten Bruch übrig bleibt, sind unabhängig. Außerdem sind die Zufallsvariablen $ Y $ und $ X/Y $ gleichverteilt auf den Mengen $[0, l]$ und $[0, 1]$, jeweils. Daher gilt $ \mathbb{E}[Y] = \frac{l}{2} $ und $ \mathbb{E}[X/Y] = \frac{1}{2} $. Somit:

$
\mathbb{E}[X] = \mathbb{E}[Y] \cdot \mathbb{E}\left[\frac{X}{Y}\right] = \frac{l}{2} \cdot \frac{1}{2} = \frac{l}{4}.
$

## Aufgabe 2: Bayes-Regel
Eine Glühbirne, die von der General Illumination Company hergestellt wird, hat eine exponentiell verteilte Lebensdauer $ Y $. Allerdings hat das Unternehmen Probleme mit der Qualitätskontrolle. An einem beliebigen Tag ist der Parameter $ \lambda $ der Dichtefunktion (PDF) von $ Y $ tatsächlich eine Zufallsvariable, die gleichverteilt im Intervall $[0, 1/2]$ ist. Wir testen eine Glühbirne und zeichnen den experimentellen Wert $ y $ ihrer Lebensdauer auf. Was können wir über den zugrunde liegenden Parameter $ \lambda $ aussagen?

Wir modellieren den Parameter $ \lambda $ als eine Zufallsvariable $ X $, die gleichverteilt ist. Alle verfügbaren Informationen über $ X $ sind in der bedingten Dichtefunktion $ f_{X|Y}(x|y) $ enthalten. Wir betrachten $ y $ als eine Konstante (gleich dem beobachteten Wert von $ Y $) und konzentrieren uns auf die Abhängigkeit der PDF von $ x $. Beachten Sie, dass $ f_X(x) = 2 $, für $ 0 \leq x \leq 1/2 $.

Gemäß der kontinuierlichen Bayes-Regel haben wir:

$
f_{X|Y}(x|y) = \frac{2xe^{-xy}}{\int_0^{1/2} 2te^{-ty}dt}, \quad \text{für } 0 \leq x \leq \frac{1}{2}.
$

In einigen Fällen ist das unbeobachtete Phänomen von Natur aus diskret. Beispielsweise, wenn ein binäres Signal in Anwesenheit von Rauschen mit einer Normalverteilung beobachtet wird. Oder wenn eine medizinische Diagnose auf der Grundlage kontinuierlicher Messungen wie Temperatur und Blutwerte gestellt wird. In solchen Fällen gilt eine etwas andere Version der Bayes-Regel.

Sei $ X $ eine diskrete Zufallsvariable, die Werte in einer endlichen Menge $\{1, \dots, n\}$ annimmt und die die verschiedenen diskreten Möglichkeiten für das unbeobachtete Phänomen von Interesse darstellt. Die Wahrscheinlichkeitsfunktion (PMF) $ p_X(x) $ von $ X $ wird als bekannt angenommen. Sei $ Y $ eine kontinuierliche Zufallsvariable, die für jeden gegebenen Wert $ x $ durch eine bedingte Dichtefunktion $ f_{Y|X}(y|x) $ beschrieben wird. Wir interessieren uns für die bedingte PMF von $ X $, gegeben den experimentellen Wert $ y $ von $ Y $.

Anstatt mit dem bedingten Ereignis $\{Y = y\}$ zu arbeiten, welches eine Null-Wahrscheinlichkeit hat, bedingen wir stattdessen das Ereignis $\{y \leq Y \leq y + \delta\}$, wobei $ \delta $ eine kleine positive Zahl ist. Dann nehmen wir den Grenzwert, wenn $ \delta $ gegen Null strebt. Mit der Bayes-Regel haben wir:

$
P(X = x | Y = y) \approx P(X = x | y \leq Y \leq y + \delta) = \frac{p_X(x)P(y \leq Y \leq y + \delta | X = x)}{P(y \leq Y \leq y + \delta)}.
$

Wir setzen dies fort:

$
\approx \frac{p_X(x)f_{Y|X}(y|x)\delta}{f_Y(y)\delta} = \frac{p_X(x)f_{Y|X}(y|x)}{f_Y(y)}.
$

Der Nenner kann unter Verwendung des Satzes der totalen Wahrscheinlichkeit ausgewertet werden. Wir haben:

$
f_Y(y) = \sum_{i=1}^n p_X(i)f_{Y|X}(y|i).
$

## Lösung:

### (a)
Wir haben:

$
\mathbb{E}[X] = \int_1^3 \frac{x^2}{4} dx = \frac{x^3}{12} \bigg|_1^3 = \frac{27}{12} - \frac{1}{12} = \frac{26}{12} = \frac{13}{6}.
$

$
P(A) = \int_2^3 \frac{x}{4} dx = \frac{x^2}{8} \bigg|_2^3 = \frac{9}{8} - \frac{4}{8} = \frac{5}{8}.
$

Wir haben außerdem:

$
f_{X|A}(x) =
\begin{cases}
\frac{f_X(x)}{P(A)}, & \text{falls } x \in A, \\
0, & \text{sonst},
\end{cases}
$

$
f_{X|A}(x) =
\begin{cases}
\frac{2x}{5}, & \text{falls } 2 \leq x \leq 3, \\
0, & \text{sonst}.
\end{cases}
$

Daraus erhalten wir:

$
\mathbb{E}[X|A] = \int_2^3 x \cdot \frac{2x}{5} dx = \frac{2x^3}{15} \bigg|_2^3 = \frac{54}{15} - \frac{16}{15} = \frac{38}{15}.
$

### (b)
Wir haben:

$
\mathbb{E}[Y] = \mathbb{E}[X^2] = \int_1^3 \frac{x^3}{4} dx = 5,
$

$
\mathbb{E}[Y^2] = \mathbb{E}[X^4] = \int_1^3 \frac{x^5}{4} dx = \frac{91}{3}.
$

Somit:

$
\text{var}(Y) = \mathbb{E}[Y^2] - (\mathbb{E}[Y])^2 = \frac{91}{3} - 5^2 = \frac{16}{3}.
$

## Aufgabe 3: Buffonsches Nadelproblem

Das Buffonsche Nadelproblem ist ein klassisches Problem der Wahrscheinlichkeitsrechnung, das von dem französischen Mathematiker Georges-Louis Leclerc de Buffon im 18. Jahrhundert formuliert wurde. Es geht darum, eine Nadel zufällig auf eine Fläche zu werfen, die mit parallelen Linien in gleichmäßigen Abständen $ t $ markiert ist, und die Wahrscheinlichkeit zu berechnen, dass die Nadel eine der Linien schneidet.

Gegeben sind folgende Bedingungen:
- Länge der Nadel: $ L $
- Abstand zwischen den parallelen Linien: $ t $

1. **Formuliere die Wahrscheinlichkeit, dass die Nadel eine der Linien schneidet, in Abhängigkeit von $ L $ und $ t $.**
2. **Wie verändert sich die Wahrscheinlichkeit, wenn die Nadellänge $ L $ gleich dem Abstand $ t $ ist?**
3. **Implementiere einen Monte-Carlo-Algorithmus, um die Wahrscheinlichkeit numerisch zu bestimmen.**

### Lösung:

1. **Wahrscheinlichkeit**:

Die Wahrscheinlichkeit, dass eine Nadel eine der Linien schneidet, ist gegeben durch die folgende Formel:

$
P = \frac{2L}{t\pi}
$

Dabei ist $ L $ die Länge der Nadel und $ t $ der Abstand zwischen den Linien. Diese Formel gilt, wenn $ L \leq t $. Falls $ L > t $, wird die Berechnung komplizierter, aber wir bleiben hier bei dem Fall $ L \leq t $.

2. **Spezialfall: $ L = t $**:

Wenn die Nadellänge gleich dem Abstand zwischen den Linien ist, also $ L = t $, dann wird die Wahrscheinlichkeit zu:

$
P = \frac{2t}{t\pi} = \frac{2}{\pi} \approx 0.6366
$

Das bedeutet, dass in etwa 63,66% der Fälle die Nadel eine der Linien schneiden wird.

3. **Monte-Carlo-Algorithmus zur numerischen Bestimmung der Wahrscheinlichkeit**:

Ein Monte-Carlo-Ansatz eignet sich gut, um die Wahrscheinlichkeit numerisch zu bestimmen. Hier ist ein Python-Skript, das diese Simulation durchführt:

```python
import numpy as np

def monte_carlo_buffon(num_simulations, L, t):
    hits = 0
    
    for _ in range(num_simulations):
        # Zufälliger Winkel zwischen 0 und pi
        theta = np.random.uniform(0, np.pi)
        # Zufällige Position des Mittelpunkts der Nadel zwischen 0 und t/2 (da symmetrisch)
        d = np.random.uniform(0, t/2)
        
        # Check if needle crosses a line
        if (L / 2) * np.sin(theta) >= d:
            hits += 1

    # Probability estimate
    return hits / num_simulations

# Parameter definition
L = 1.0  # Länge der Nadel
t = 1.0  # Abstand zwischen den Linien
num_simulations = 1_000_000  # Anzahl der Simulationen

# Monte-Carlo Simulation aufrufen
probability_estimate = monte_carlo_buffon(num_simulations, L, t)
print(f"Geschätzte Wahrscheinlichkeit: {probability_estimate}")
```

## Aufgabe 4: Maximum-Likelihood-Schätzung

Gegeben sei eine Stichprobe von $n$ Werten $ X = \{x_1, x_2, \dots, x_n\} $, die einer Exponentialverteilung mit dem Parameter $\lambda$ folgt. Die Dichtefunktion der Exponentialverteilung ist:

$
f(x|\lambda) = \lambda e^{-\lambda x} \quad \text{für} \, x \geq 0
$

1. **Leite die Maximum-Likelihood-Schätzung für den Parameter $ \lambda $ her.**
2. **Berechne die Schätzung von $ \lambda $, wenn die Stichprobe $ X = \{2, 3, 1, 4, 5\} $ gegeben ist.**

### Lösung:

1. **Maximum-Likelihood-Schätzung**:

Die Likelihood-Funktion für eine Stichprobe von $ n $ unabhängigen Beobachtungen $ x_1, x_2, \dots, x_n $ ist das Produkt der Wahrscheinlichkeitsdichten:

$
L(\lambda|X) = \prod_{i=1}^n \lambda e^{-\lambda x_i}
$

Das Log-Likelihood wird dann:

$
\log L(\lambda|X) = n \log(\lambda) - \lambda \sum_{i=1}^n x_i
$

Die Maximum-Likelihood-Schätzung (MLE) wird durch Maximierung des Log-Likelihoods nach $ \lambda $ gefunden. Dazu leiten wir das Log-Likelihood nach $ \lambda $ ab und setzen es gleich Null:

$
\frac{\partial \log L(\lambda|X)}{\partial \lambda} = \frac{n}{\lambda} - \sum_{i=1}^n x_i = 0
$

Dies führt zur Schätzung:

$
\hat{\lambda} = \frac{n}{\sum_{i=1}^n x_i}
$

2. **Schätzung von $ \lambda $ für die Stichprobe $ X = \{2, 3, 1, 4, 5\} $**:

Die Summe der Stichprobenwerte beträgt:

$
\sum_{i=1}^n x_i = 2 + 3 + 1 + 4 + 5 = 15
$

Die Schätzung von $ \lambda $ ist dann:

$
\hat{\lambda} = \frac{5}{15} = \frac{1}{3} \approx 0,333
$
