# Multinomialverteilung

```r
# Anzahl der Elemente und Wahrscheinlichkeiten definieren
anzahl <- 3
wahrscheinlichkeiten <- c(0.5, 0.2, 0.3)

# Matrix 
D <- matrix(c(1, 3, 5, 1, 2, 4, 1, 3, 3, 1, 4, 2, 
              1, 5, 1, 2, 0, 4, 2, 0, 5, 2, 3, 1, 
              3, 0, 3, 1, 4, 1, 2, 0, 4, 1, 1, 2, 
              5, 0, 1, 5, 1, 0, 6, 3, 2), 
            byrow = FALSE, nrow = 3)

# Berechnung der Multinomialverteilung
W <- rep(NA, 15)
for (j in 1:15) W[j] <- dmultinom(D[, j], prob = wahrscheinlichkeiten)

# Ergebnisse auf drei Dezimalstellen runden
runde_W <- round(W, 3)
print(runde_W)

# Scatterplot3D erstellen
library(scatterplot3d)                                   
par(mfrow = c(1, 1), lwd = 2, font.axis = 2, bty = "n", ps = 12)
z_werte <- W
x_werte <- D[1,]
y_werte <- D[2,]

scatterplot3d(x_werte, y_werte, z_werte, col.axis = "black", type = "h", 
               highlight.3d = TRUE, box = FALSE, col.grid = "grey", 
               pch = 20, cex.symbols = 2, 
               xlab = "x (Anzahl fÃ¼r Gruppe 1)", 
               ylab = "y (Anzahl fÃ¼r Gruppe 2)", 
               zlab = "P(X=x, Y=y)")
```
