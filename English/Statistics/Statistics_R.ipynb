# Discrete Uniform Distribution using the Example of a Die

```r
# Set the layout for the plots
par(mfrow = c(1, 2), lwd = 2, font.axis = 2, bty = "n", ps = 11)

# Define new variables
numbers <- seq(1, 6, by = 1)          # Die numbers
probabilities <- rep(1/6, 6)          # Probability distribution

# First plot for the probability distribution f(x)
plot(numbers, probabilities, type = "h", ylim = c(0, 0.2), 
     ylab = "Probability f(x)", xlab = "Die roll", col = "lightblue")
points(numbers, probabilities, pch = 19, cex = 1.1, col = "black")

# Define new x-values for the cumulative plot
new_x <- c(0, numbers, 7)  
cumulative_F <- c(0, cumsum(probabilities), 1)

# Second plot for the cumulative distribution function F(x)
plot(new_x, cumulative_F, type = "s", ylim = c(0, 1), 
     ylab = "Cumulative Distribution F(x)", xlab = "Die roll")
```

# Multinomial Distribution

```r
# Define number of elements and probabilities
count <- 3
probabilities <- c(0.5, 0.2, 0.3)

# Matrix
D <- matrix(c(1, 3, 5, 1, 2, 4, 1, 3, 3, 1, 4, 2, 
              1, 5, 1, 2, 0, 4, 2, 0, 5, 2, 3, 1, 
              3, 0, 3, 1, 4, 1, 2, 0, 4, 1, 1, 2, 
              5, 0, 1, 5, 1, 0, 6, 3, 2), 
            byrow = FALSE, nrow = 3)

# Calculation of the multinomial distribution
W <- rep(NA, 15)
for (j in 1:15) W[j] <- dmultinom(D[, j], prob = probabilities)

# Round results to three decimal places
rounded_W <- round(W, 3)
print(rounded_W)

# Create 3D Scatterplot
library(scatterplot3d)                                   
par(mfrow = c(1, 1), lwd = 2, font.axis = 2, bty = "n", ps = 12)
z_values <- W
x_values <- D[1,]
y_values <- D[2,]

scatterplot3d(x_values, y_values, z_values, col.axis = "black", type = "h", 
               highlight.3d = TRUE, box = FALSE, col.grid = "grey", 
               pch = 20, cex.symbols = 2, 
               xlab = "x (Count for Group 1)", 
               ylab = "y (Count for Group 2)", 
               zlab = "P(X=x, Y=y)")
```
